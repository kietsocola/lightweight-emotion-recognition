```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Danh sách tên ảnh từ 01.jpg đến 07.jpg
image_filenames = [f'{i:02}.jpg' for i in range(1, 8)]

# Lặp qua từng ảnh
for filename in image_filenames:
    # Đọc ảnh và chuyển sang RGB
    image = cv2.imread(filename)
    if image is None:
        print(f"Không tìm thấy ảnh: {filename}")
        continue

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Tạo ảnh tối bằng cách giảm độ sáng
    dark_image = (image * 0.3).astype(np.uint8)

    # Hiển thị ảnh tối
    plt.figure(figsize=(5, 5))
    plt.imshow(dark_image)
    plt.title(f'{filename} - Ảnh tối')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```
