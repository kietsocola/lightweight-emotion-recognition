```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Danh sch tn nh t 01.jpg n 07.jpg
image_filenames = [f'{i:02}.jpg' for i in range(1, 8)]

# Lp qua tng nh
for filename in image_filenames:
    # c nh v chuyn sang RGB
    image = cv2.imread(filename)
    if image is None:
        print(f"Khng tm thy nh: {filename}")
        continue

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # To nh ti bng cch gim  sng
    dark_image = (image * 0.3).astype(np.uint8)

    # Hin th nh ti
    plt.figure(figsize=(5, 5))
    plt.imshow(dark_image)
    plt.title(f'{filename} - nh ti')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```
