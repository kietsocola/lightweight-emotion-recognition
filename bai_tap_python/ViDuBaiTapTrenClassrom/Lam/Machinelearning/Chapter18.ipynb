### 18.1 Thực Hành Học Máy Với Các Dự Án

Làm việc qua các vấn đề học máy từ đầu đến cuối là cực kỳ quan trọng. Bạn có thể đọc về học máy. Bạn cũng có thể thử các công thức nhỏ lẻ. Nhưng học máy ứng dụng sẽ không trở nên sống động với bạn cho đến khi bạn làm việc qua một tập dữ liệu từ đầu đến cuối.

Làm việc qua một dự án buộc bạn phải suy nghĩ về cách mô hình sẽ được sử dụng, thách thức các giả định của bạn và rèn luyện tốt ở tất cả các phần của một dự án, không chỉ những phần bạn yêu thích. Cách tốt nhất để thực hành các dự án dự đoán học máy là sử dụng các tập dữ liệu chuẩn hóa từ Kho Lưu Trữ Học Máy UCI. Một khi bạn có một tập dữ liệu thực hành và một loạt công thức Python, làm thế nào để bạn kết hợp tất cả lại và làm việc qua vấn đề từ đầu đến cuối?

#### 18.1.1 Sử Dụng Quy Trình Có Cấu Trúc Theo Bước

Bất kỳ dự án học máy dự đoán nào cũng có thể được chia thành sáu nhiệm vụ chung:

1. Xác Định Vấn Đề.
2. Tóm Tắt Dữ Liệu.
3. Chuẩn Bị Dữ Liệu.
4. Đánh giá thuật toán.

6. Trình bày kết quả.

Các kỹ năng có thể kết hợp hoặc bỏ qua tùy thuộc vào vấn đề, nhưng đây là cấu trúc tổng quát để làm việc qua các vấn đề học máy dự đoán trong Python. Bạn cần phải ánh xạ các bước này thành các công thức học máy (ví dụ: Pandas cho việc tải dữ liệu và scikit-learn cho việc mô hình hóa). Phần tiếp theo cung cấp chi tiết về cách ánh xạ này và giải thích từng nhiệm vụ cùng các loại tiểu nhiệm vụ mà các thư viện có thể sử dụng.

### 18.2 Machine Learning Project Template in Python

#### 18.2.1 Template Summary

```python
# Python Project Template
# 1. Prepare Problem
# a) Load libraries
# b) Load dataset

# 2. Summarize Data
# a) Descriptive statistics
# b) Data visualizations

# 3. Prepare Data
# a) Data Cleaning
# b) Feature Selection
# c) Data Transforms

# 4. Evaluate Algorithms
# a) Split-out validation dataset
# b) Test options and evaluation metric
# c) Spot Check Algorithms
# d) Compare Algorithms

# 5. Improve Accuracy
# a) Algorithm Tuning
# b) Ensembles

# 6. Finalize Model
# a) Predictions on validation dataset
# b) Create standalone model on entire training dataset
# c) Save model for later use
```

## 18.3 Machine Learning Project Template Steps

Phần này cung cấp cho bạn thêm chi tiết về từng bước của mẫu.

#### 18.3.1 Prepare Problem

Bước này nói về việc chuẩn bị mọi thứ cần thiết để bắt đầu làm việc trên vấn đề của bạn. Điều này bao gồm:
- Các mô-đun Python, lớp và hàm mà bạn dự định sử dụng.
- Tải dữ liệu của bạn từ CSV.

Đây là bước đầu tiên của một quy trình cấu hình toàn cầu mà bạn có thể cần thực hiện. Nó cũng là nơi lý tưởng để làm rõ vấn đề của bạn và đảm bảo rằng nó không quá lớn để xử lý trong một lần, thường là 30-60 phút. Bạn luôn có thể mở rộng các mô hình hoạt động tốt sau này.

#### 18.3.2 Summarize Data

Bước này là về việc hiểu tốt hơn dữ liệu mà bạn có sẵn. Điều này bao gồm việc sử dụng dữ liệu:
- Thống kê mô tả như tóm tắt.
- Trực quan hóa dữ liệu như biểu đồ với Matplotlib, lý tưởng là sử dụng các hàm tiện lợi từ Pandas.

Thời gian này có thể sử dụng để đặt ra nhiều câu hỏi, giả định và giả thuyết mà bạn có thể điều tra sau này với các mô hình chuyên biệt.

#### 18.3.3 Prepare Data

Bước này là về việc chuẩn bị dữ liệu theo cách tốt nhất để thể hiện cấu trúc của vấn đề và các mối quan hệ giữa các thuộc tính đầu vào với biến đầu ra. Điều này bao gồm các nhiệm vụ như:
- Làm sạch dữ liệu bằng cách xóa trùng lặp, đánh dấu giá trị bị thiếu và thậm chí nhập giá trị bị thiếu.
- Lựa chọn đặc trưng nơi các đặc trưng dư thừa có thể bị loại bỏ và các đặc trưng mới được phát triển.
- Chuyển đổi dữ liệu nơi các thuộc tính được chuẩn hóa hoặc phân phối lại theo thứ tự để tốt nhất thể hiện cấu trúc của vấn đề sau này khi học thuật toán.

Bắt đầu đơn giản. Xem lại bước này thường xuyên và lặp lại với bước tiếp theo cho đến khi bạn đạt được một biểu diễn của dữ liệu dẫn đến các mô hình chính xác hoặc đủ chính xác để tiếp tục.

#### 18.3.4 Evaluate Algorithms

Bước này nhằm tìm ra một tập hợp các thuật toán học máy tốt nhất để khai thác cấu trúc của dữ liệu của bạn, thường tốt hơn so với việc học ngẫu nhiên. Điều này bao gồm các bước như:
- Tách một tập dữ liệu xác thực để sử dụng sau này nhằm xác nhận kỹ năng của bạn.
- Xây dựng các tùy chọn kiểm tra bằng scikit-learn như kiểm tra chéo và thước đo đánh giá.
- Sử dụng.
- So sánh độ chính xác ước tính của các thuật toán.

Trên một vấn đề cụ thể, bạn sẽ dành phần lớn thời gian của mình cho bước này và bước trước đó cho đến khi bạn tìm thấy một tập hợp 3-5 thuật toán học máy hoạt động tốt.

#### 18.3.5 Improve Accuracy

Một khi bạn có một danh sách ngắn các thuật toán học máy, bạn cần phải tối ưu hóa chúng. Có hai cách khác nhau để cải thiện độ chính xác của mô hình của bạn:
- Tìm kiếm cho một tổ hợp các tham số tối ưu cho mỗi thuật toán sử dụng scikit-learn mà mang lại kết quả tốt nhất.
- Kết hợp dự đoán của nhiều mô hình thành một mô hình dự đoán tổng hợp bằng cách sử dụng ensemble.

Có thể có một chút điều chỉnh thuật toán trong bước trước đó. Và trong trường hợp ensemble, bạn có thể có nhiều hơn một danh sách ngắn các thuật toán để kết hợp dự đoán của chúng.

#### 18.3.6 Finalize Model

Một khi bạn đã tìm thấy một mô hình mà bạn tin rằng có thể đưa ra dự đoán chính xác trên dữ liệu chưa được nhìn thấy, bạn đã sẵn sàng để hoàn thiện nó. Hoàn thiện một mô hình có thể bao gồm các tiểu nhiệm vụ như:
- Sử dụng mô hình tối ưu đã được tinh chỉnh bằng scikit-learn để đưa ra dự đoán trên dữ liệu chưa được nhìn thấy.
- Tạo một mô hình độc lập sử dụng các tham số đã được tinh chỉnh bằng scikit-learn.
- Lưu mô hình tối ưu để sử dụng sau này.

Khi bạn đã hoàn thành đến bước này, bạn đã sẵn sàng để trình bày kết quả cho các bên liên quan và/hoặc triển khai mô hình của mình để bắt đầu đưa ra dự đoán trên dữ liệu chưa được nhìn thấy.
