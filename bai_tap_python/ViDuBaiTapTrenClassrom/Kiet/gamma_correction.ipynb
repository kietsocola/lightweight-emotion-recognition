## 1Ô∏è‚É£ Gamma l√† g√¨?  

Gamma (Œ≥) l√† m·ªôt tham s·ªë ki·ªÉm so√°t ƒë·ªô s√°ng v√† ƒë·ªô t∆∞∆°ng ph·∫£n c·ªßa h√¨nh ·∫£nh b·∫±ng c√°ch thay ƒë·ªïi c√°ch √°nh s√°ng ƒë∆∞·ª£c √°nh x·∫° t·ª´ ƒë·∫ßu v√†o sang ƒë·∫ßu ra.  

### üîπ C√¥ng th·ª©c Gamma Correction:  
$$
I_{\text{new}} = 255 \times \left( \frac{I_{\text{old}}}{255} \right)^{\frac{1}{\gamma}}
$$

- $I_{\text{old}}$ l√† gi√° tr·ªã pixel ban ƒë·∫ßu (0-255).  
- $I_{\text{new}}$ l√† gi√° tr·ªã pixel sau khi √°p d·ª•ng gamma correction.  
- $\gamma$ l√† gi√° tr·ªã gamma.

---

## 2Ô∏è‚É£ ·∫¢nh h∆∞·ªüng c·ªßa Gamma Correction  

- N·∫øu **$\gamma > 1$** ‚Üí ·∫¢nh s·∫Ω **s√°ng h∆°n**.  
- N·∫øu **$\gamma < 1$** ‚Üí ·∫¢nh s·∫Ω **t·ªëi h∆°n**.  

V√≠ d·ª•:  

| Gi√° tr·ªã Gamma | ·∫¢nh ƒë·∫ßu ra |
|--------------|------------|
| **$\gamma = 0.5$** | ·∫¢nh t·ªëi h∆°n |
| **$\gamma = 1$** | Kh√¥ng thay ƒë·ªïi |
| **$\gamma = 2.2$** | ·∫¢nh s√°ng h∆°n |

---

## 3Ô∏è‚É£ Code Python minh h·ªça Gamma Correction  

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# H√†m ƒëi·ªÅu ch·ªânh Gamma Correction
def adjust_gamma(image, gamma=1.0):
    inv_gamma = 1.0 / gamma
    table = np.array([(i / 255.0) ** inv_gamma * 255 for i in range(256)]).astype("uint8")
    return cv2.LUT(image, table)

# ƒê·ªçc ·∫£nh g·ªëc (c·∫ßn c√≥ s·∫µn file ·∫£nh)
image = cv2.imread("dark_image.png")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Chuy·ªÉn ·∫£nh t·ª´ BGR sang RGB ƒë·ªÉ hi·ªÉn th·ªã ƒë√∫ng m√†u

# √Åp d·ª•ng Gamma Correction
gamma_corrected = adjust_gamma(image, gamma=2.2)

# Hi·ªÉn th·ªã ·∫£nh g·ªëc v√† ·∫£nh ƒë√£ x·ª≠ l√Ω
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("·∫¢nh G·ªëc")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected)
plt.title("·∫¢nh Sau Gamma Correction")
plt.axis("off")

plt.show()
```
